```python
import pandas as pd
import altair as alt

df_di = pd.read_excel('Resources/portafolio_innova.xlsx')

# esto está hecho a mano por que no hay metodo de comparación empírica entre el sector y las areas ocde 
# o anid descritas en el documento

def SetAreaOCDE(sector):
    if sector == 'Agrícola (excepto vitivinicola)':
        return 'Ciencias Agrícolas'
    elif sector == 'Alimentos (excepto vitivinicola)':
        return 'Ingeniería, Tecnología ...'
    elif sector == 'Servicios de ingeniería o de conocimiento':
        return 'Ingeniería, Tecnología ...'
    elif sector == 'Manufactura de metales básicos':
        return 'Ingeniería, Tecnología ...'
    elif sector == 'Química, caucho y plásticos (excepto industria farmaceutica)':
        return 'Ingeniería, Tecnología ...'
    elif sector == 'Minería y metalurgia extractiva':
        return 'Ingeniería, Tecnología ...'
    elif sector == 'Construcción':
        return 'Ingeniería, Tecnología ...'
    elif sector == 'Farmaceutica':
        return 'Ciencias médicas, Ciencias de la salud'
    elif sector == 'Comercio y retail':
        return 'Ciencias Sociales'
    elif sector == 'Pesca y acuicultura':
        return 'Ciencias Agrícolas'
    elif sector == 'Servicios empresariales administrativos y de apoyo':
        return 'Ciencias Sociales'
    elif sector == 'Forestal':
        return 'Ciencias Agrícolas'
    elif sector == 'Gestión de desechos':   #NO SÉ COMO CLASIFICARLO MEJOR
        return 'Gestión de desechos'
    elif sector == 'Educación':
        return 'Ciencias Sociales'
    elif sector == 'Ganadero':
        return 'Ciencias Agrícolas'
    elif sector == 'Manufactura de maquinaria y equipos (Metalmecánico)':
        return 'Ingeniería, Tecnología ...'
    elif sector ==  'Tecnologías de la información':
        return 'Ingeniería, Tecnología ...'       #PODRIA QUEDAR COMO CIENCIAS NATURALES
                                              # según: https://vrid.udec.cl/sites/default/files/Areas%20de%20Investigación%20OCDE_0.pdf#overlay-context=formularioseinstructivos
    elif sector == 'Salud y asistencia social':
        return 'Ciencias médicas, Ciencias de la salud'        
    elif sector == 'Logística y Transporte':
        return 'Ingeniería, Tecnología ...' 
    elif sector == 'Recursos Hídricos':
        return 'Ciencias Naturales'
    elif sector == 'Manufactura de no metálicos':
        return 'Ingeniería, Tecnología ...'
    elif sector == 'Energético':
        return 'Ciencias Naturales'
    elif sector == 'Sector público':
        return 'Ciencias Sociales'
    elif sector =='Banca y sector financiero': 
        return 'Ciencias Sociales'
    elif sector == 'Telecomunicaciones':
        return 'Ingeniería, Tecnología ...'
    elif sector == 'Inmobiliario':
        return 'Ingeniería, Tecnología ...'
    elif sector == 'Industria creativa y de esparcimiento':
        return 'Ciencias Sociales'
    elif sector == 'Asociaciones y organizaciones no empresariales ni gubernamentales':
        return 'Ciencias Sociales'
    elif sector == 'Turismo':
        return 'Ciencias Sociales'
    elif sector == 'Otras industrias manufactureras':
        return 'Ingeniería, Tecnología ...'
    return sector

def SetAreaANID(sector):
    if sector == 'Agrícola (excepto vitivinicola)':
        return 'Tecnología, Agrícola y Productividad'
    elif sector == 'Alimentos (excepto vitivinicola)':
        return 'Tecnología y ciencias Silvoagropecuarias'
    elif sector == 'Servicios de ingeniería o de conocimiento':
        return 'Tecnología y ciencias de la Ingeniería'
    elif sector == 'Manufactura de metales básicos':
        return 'Tecnología y ciencias de la Ingeniería'
    elif sector == 'Química, caucho y plásticos (excepto industria farmaceutica)':
        return 'Ciencias exactas y naturales'
    elif sector == 'Minería y metalurgia extractiva':
        return 'Ciencias exactas y naturales'
    elif sector == 'Construcción':
        return 'Tecnología y ciencias de la Ingeniería'
    elif sector == 'Farmaceutica':
        return 'Ciencias exactas y naturales'
    elif sector == 'Comercio y retail':
        return 'Ciencias jurídicas, económicas y administrativas'
    elif sector == 'Pesca y acuicultura':
        return 'Tecnología y ciencias Silvoagropecuarias'
    elif sector == 'Servicios empresariales administrativos y de apoyo':
        return 'Ciencias jurídicas, económicas y administrativas'
    elif sector == 'Forestal':
        return 'Tecnología y ciencias Silvoagropecuarias'
    elif sector == 'Gestión de desechos':  
        return 'Ciencias sociales'
    elif sector == 'Educación':
        return 'Ciencias sociales'
    elif sector == 'Ganadero':
        return 'Tecnología y ciencias Silvoagropecuarias'
    elif sector == 'Manufactura de maquinaria y equipos (Metalmecánico)':
        return 'Tecnología y ciencias de la Ingeniería'
    elif sector ==  'Tecnologías de la información':
        return  'Tecnología y ciencias de la Ingeniería'
    elif sector == 'Salud y asistencia social':
        return  'Tecnología y ciencias médicas'
    elif sector == 'Logística y Transporte':
        return 'Tecnología y ciencias de la Ingeniería'
    elif sector == 'Recursos Hídricos':
        return 'Ciencias exactas y naturales'
    elif sector == 'Manufactura de no metálicos':
        return 'Tecnología y ciencias de la Ingeniería'
    elif sector == 'Energético':
        return 'Ciencias jurídicas, económicas y administrativas'
    elif sector == 'Sector público':
        return 'Ciencias jurídicas, económicas y administrativas'
    elif sector =='Banca y sector financiero': 
        return 'Ciencias jurídicas, económicas y administrativas'
    elif sector == 'Telecomunicaciones':
        return 'Tecnología y ciencias de la Ingeniería'
    elif sector == 'Inmobiliario':
        return 'Tecnología y ciencias de la Ingeniería'
    elif sector == 'Industria creativa y de esparcimiento':
        return 'Ciencias sociales'
    elif sector == 'Asociaciones y organizaciones no empresariales ni gubernamentales':
        return 'Ciencias sociales'
    elif sector == 'Turismo':
        return 'Ciencias sociales'
    elif sector == 'Otras industrias manufactureras':
        return 'Tecnología y ciencias de la Ingeniería'
    return sector

```

```python
#Nos centraremos en el foco de desarrollo innovación con I+D
df_id = df_di[df_di['foco_apoyo']=='Desarrolla innovación con I+D'].copy()
df_id['area_ocde'] = df_id['sector_equivalente'].apply(SetAreaOCDE)
df_id['area_anid'] = df_id['sector_equivalente'].apply(SetAreaANID)
#Incorporamos normalización por habitantes
population = [
    {"regionejecucion": "Región del Libertador General Bernardo O'Higgins", "Population": 914555},
    {"regionejecucion": "Región del Maule", "Population": 1044950},
    {"regionejecucion": "Región de Ñuble", "Population": 480609},
    {"regionejecucion": "Región del Biobío", "Population": 1556805}, 
    {"regionejecucion": "Región Metropolitana de Santiago", "Population": 7112808}
] # Fuente: Instituto Nacional de Estadistica
df_population = pd.DataFrame(population)
#df_id = df_id.merge(df_population, on="regionejecucion") #la idea es hacer este merge cuando se aplique agrupamiento por región
df_id.head(3)
```

# 5. ¿En qué áreas se está invirtiendo más en la región los últimos 3 años?

```python
# seteamos los últimos 3 años
df_id = df_id[df_id['anio_adjudicacion'].isin([2019,2020,2021])].copy()
# Comenzamos a trabajar con Biobio y RM ---- Se pueden seguir agregando regiones para explorar visualizaciones
df_id = df_id[df_id['regionejecucion'].isin(['Región del Biobío','Región Metropolitana de Santiago','Región de Ñuble',
                                             'Región del Maule',"Región del Libertador General Bernardo O'Higgins"])]

# Agrupamos por 
df_pg5 = df_id.copy()
df_pg5 = df_pg5.groupby(['sector_equivalente','regionejecucion','anio_adjudicacion'])\
               .agg({'aprobado_corfo':'sum','aprobado_privado':'sum','monto_consolidado_ley':'sum'})\
               .reset_index()
df_pg5 = df_pg5.merge(df_population, on="regionejecucion")
#tidy data para los montos
df_pg5 = pd.melt(frame=df_pg5, id_vars=['sector_equivalente','regionejecucion','Population'],
                   value_vars=['aprobado_privado','aprobado_corfo','monto_consolidado_ley'],
                  var_name='financiamiento', value_name='monto')
#normalizamos 
# millones de pesos por cada 100mil hb
df_pg5['monto'] = (df_pg5['monto'] * 100000 )/(df_pg5['Population']*1000000)  
#esta linea va para evitar que se vean lineas de unión en el gráfico
df_pg5 = df_pg5.groupby(['sector_equivalente','regionejecucion','financiamiento'])\
            .agg({'monto':'sum'}).reset_index()
df_pg5.head()
```

```python
# Gráfico lo consideramos con información de data innovación -- sector Equivalente
chart_pg5A = alt.Chart(df_pg5).mark_bar().encode(
                y = alt.Y('sector_equivalente:O',title=None, sort='-x'),
                x = alt.X('monto:Q', title='M$ por cada 100 mil hb'),
                column = alt.Column('regionejecucion',title=None),
                color = alt.Color('financiamiento', title='Tipo de Financiamiento',scale=alt.Scale(scheme='category10')),
                tooltip = alt.Tooltip('monto:Q')
            )
chart_pg5A
```

```python
# PARA AREA OCDE
df_pg5B = df_id.copy()
df_pg5B = df_pg5B.groupby(['area_ocde','regionejecucion','anio_adjudicacion'])\
               .agg({'aprobado_corfo':'sum','aprobado_privado':'sum','monto_consolidado_ley':'sum'})\
               .reset_index()
df_pg5B = df_pg5B.merge(df_population, on="regionejecucion")
#tidy data para los montos
df_pg5B = pd.melt(frame=df_pg5B, id_vars=['area_ocde','regionejecucion','Population'],
                   value_vars=['aprobado_privado','aprobado_corfo','monto_consolidado_ley'],
                  var_name='financiamiento', value_name='monto')
#normalizamos 
df_pg5B['monto'] = (df_pg5B['monto'] * 100000 )/(df_pg5B['Population']*1000000)  
df_pg5B = df_pg5B.groupby(['area_ocde','regionejecucion','financiamiento'])\
            .agg({'monto':'sum'}).reset_index()
df_pg5B.head()
```

```python
# Gráfico lo consideramos con información de data innovación -- AREA OCDE
chart_pg5B = alt.Chart(df_pg5B).mark_bar().encode(
                y = alt.Y('area_ocde:O',title=None, sort='-x'),
                x = alt.X('monto:Q', title='M$ por cada 100 mil hb'),
                column = alt.Column('regionejecucion',title=None),
                color = alt.Color('financiamiento', title='Tipo de Financiamiento',scale=alt.Scale(scheme='category10')),
            )
chart_pg5B
```

```python
# PARA AREA ANID
df_pg5C = df_id.copy()
df_pg5C = df_pg5C.groupby(['area_anid','regionejecucion','anio_adjudicacion'])\
               .agg({'aprobado_corfo':'sum','aprobado_privado':'sum','monto_consolidado_ley':'sum'})\
               .reset_index()
df_pg5C = df_pg5C.merge(df_population, on="regionejecucion")
#tidy data para los montos
df_pg5C = pd.melt(frame=df_pg5C, id_vars=['area_anid','regionejecucion','Population'],
                   value_vars=['aprobado_privado','aprobado_corfo','monto_consolidado_ley'],
                  var_name='financiamiento', value_name='monto')
#normalizamos 
df_pg5C['monto'] = (df_pg5C['monto'] * 100000 )/(df_pg5C['Population']*1000000 )
df_pg5C = df_pg5C.groupby(['area_anid','regionejecucion','financiamiento'])\
            .agg({'monto':'sum'}).reset_index()
df_pg5C.head()
```

```python
# Gráfico lo consideramos con información de data innovación -- AREA ANID
chart_pg5C = alt.Chart(df_pg5C).mark_bar().encode(
                y = alt.Y('area_anid:O',title=None,sort='-x'),
                x = alt.X('monto:Q', title='M$ por cada 100 mil hb'),
                column = alt.Column('regionejecucion',title=None),
                color = alt.Color('financiamiento', title='Tipo de Financiamiento',scale=alt.Scale(scheme='category10')),
            )
chart_pg5C
```

# 6. ¿En qué áreas se está observando un incremento en la producción científica en los últimos 3 años?

```python
df_pg6 = df_id.groupby(['sector_equivalente','regionejecucion']).agg({'codigo':'count'}).copy().reset_index()
#Normalizamos por cada 100mil habitantes
df_pg6 = df_pg6.merge(df_population, on="regionejecucion")
df_pg6['codigo'] = (df_pg6['codigo']/df_pg6['Population'])*100000

chart_pg6 = alt.Chart(df_pg6).mark_bar().encode(
                y = alt.Y('sector_equivalente:O',title=None, sort='-x'),
                x = alt.X('codigo:Q', title='N° proyectos por cada 100mil hb'),
                column = alt.Column('regionejecucion',title=None) 
            ).configure_mark(
                color='navy'
            )
chart_pg6
```

```python
# por área ocde
df_pg6B = df_id.groupby(['area_ocde','regionejecucion']).agg({'codigo':'count'}).copy().reset_index()
#Normalizamos por cada 100mil habitantes
df_pg6B = df_pg6B.merge(df_population, on="regionejecucion")
df_pg6B['codigo'] = (df_pg6B['codigo']/df_pg6B['Population'])*100000


chart_pg6B = alt.Chart(df_pg6B).mark_bar().encode(
                y = alt.Y('area_ocde:O',title=None, sort='-x'),
                x = alt.X('codigo:Q', title='N° proyectos por cada 100mil hb'),
                column = alt.Column('regionejecucion',title=None) 
            ).configure_mark(
                color='maroon'
            )
chart_pg6B
```

```python
# por área ocde
df_pg6C = df_id.groupby(['area_anid','regionejecucion']).agg({'codigo':'count'}).copy().reset_index()
#Normalizamos por cada 100mil habitantes
df_pg6C = df_pg6C.merge(df_population, on="regionejecucion")
df_pg6C['codigo'] = (df_pg6C['codigo']/df_pg6C['Population'])*100000

chart_pg6C = alt.Chart(df_pg6C).mark_bar().encode(
                y = alt.Y('area_anid:O',title=None, sort='-x'),
                x = alt.X('codigo:Q', title='N° proyectos por cada 100mil hb'),
                column = alt.Column('regionejecucion',title=None) 
            ).configure_mark(
                color='seagreen'
            )
chart_pg6C
```
